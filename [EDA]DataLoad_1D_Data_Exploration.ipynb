```python
# Define start and end dates
start_date = pd.to_datetime("2024-08-01")
end_date = pd.to_datetime("2025-01-01")

# Create an empty list to store DataFrames
df_list = []

# Loop through each month from Aug 2024 to Jan 2025
current_date = start_date
while current_date <= end_date:
    year_month = current_date.strftime("%Y%m")  # Format YYYYMM

    # Construct the file path
    file_path = f"extracted_csv_demand/PUBLIC_ARCHIVE#DEMANDOPERATIONALACTUAL#FILE01#{year_month}010000.CSV"

    # Check if file exists before reading
    if os.path.exists(file_path):
        df_temp = pd.read_csv(file_path, skiprows=1)  # Skip the first row
        df_list.append(df_temp)
        print(f"✅ Loaded data from {file_path}, rows: {len(df_temp)}")
    else:
        print(f"⚠️ File not found: {file_path}")

    # Move to the next month
    current_date += pd.DateOffset(months=1)

# Concatenate all DataFrames
if df_list:
    df_demand = pd.concat(df_list, ignore_index=True)
    print(f"🎉 Successfully concatenated {len(df_demand)} rows from all files!")
else:
    df_demand = pd.DataFrame()  # Empty DataFrame if no files were found
    print("❌ No data files found for the given period.")

# Display the first few rows
df_demand.head()
```

# Query data 

```python
sql_scada = """SELECT * FROM DISPATCH_UNIT_SCADA
WHERE SETTLEMENTDATE LIKE '2024/08%' OR SETTLEMENTDATE LIKE '2024/09%' 
OR SETTLEMENTDATE LIKE '2024/10%' OR SETTLEMENTDATE LIKE '2024/11%'
OR SETTLEMENTDATE LIKE '2024/12%' OR SETTLEMENTDATE LIKE '2025/01%';
"""

sql_summary = """
SELECT * FROM DUDETAILSUMMARY
WHERE START_DATE LIKE '2024/08%' OR START_DATE LIKE '2024/09%' 
   OR START_DATE LIKE '2024/10%' OR START_DATE LIKE '2024/11%' 
   OR START_DATE LIKE '2024/12%' OR START_DATE LIKE '2025/01%';
"""
```

```python
# Read data into Pandas DataFrames
df_scada = pd.read_sql_query(sql_scada, conn)
df_summary = pd.read_sql_query(sql_summary, conn)

# Close database connection
conn.close()

# Display the number of rows retrieved
print(f"✅ Loaded {len(df_scada)} rows from DISPATCH_UNIT_SCADA")
print(f"✅ Loaded {len(df_summary)} rows from DUDETAILSUMMARY")
```

#### stats

<div style="background-color: lightblue; padding: 10px;">

TRANSMISSIONLOSSFACTOR is energy losses during transmission in the grid. 

DISTRIBUTIONLOSSFACTOR is column represents the factor used to account for energy losses that occur during the distribution of electricity from the generation source to the end consumers.

MINIMUM_ENERGY_PRICE is the minimum price for energy that a unit can offer, adjusted for transmission and distribution loss factors. But they have negative values. It could represent a situation where the energy price is pushed below zero to balance the market.

MIN_RAMP_RATE_UP is the minimum rate at which a unit can increase its output (in MW/min) when transitioning from a low to a high generation state. It indicates how quickly a generator can respond when there is an increased demand for electricity or when it is asked to increase its generation output.

MAX_RAMP_RATE_DOWN is the maximum rate at which a unit can decrease its output (in MW/min).

LOAD_MINIMUM_ENERGY_PRICE also has negative values.

</div>
